```python
from shapely.geometry import Point
from shapely.geometry.polygon import Polygon
import numpy as np
```

Year 2001

```python
pixelYield_2001=[]
for j in range(0,24,1):
  x1=qgislatlong[j][1]
  y1=qgislatlong[j][2]
  x2=qgislatlong[j][3]
  y2=qgislatlong[j][4]
  x3=qgislatlong[j][5]
  y3=qgislatlong[j][6]
  x4=qgislatlong[j][7]
  y4=qgislatlong[j][8]
  farmPoly = Polygon(np.array([[x1,y1],[x2,y2],[x3,y3],[x4,y4]]))
  nrow=numberRC[j][3]
  ncol=numberRC[j][2]
  yd=(int(nrow),int(ncol))
  yd=np.zeros(yd)
  countyd=(int(nrow),int(ncol))
  countyd = np.zeros(countyd)
  horizontaldiff_x = (x2-x1)/ncol;
  verticaldiff_y = (y4-y1)/nrow;
  ans=(int(nrow),int(ncol))
  ans=np.zeros(ans)
  for i in range(0,int(X_2001.size/X_2001[0].size),1):
    point=Point(X_2001[i][0],X_2001[i][1])
    if point.within(farmPoly):
      for r in range(0,int(nrow),1):
        for c in range(0,int(ncol),1):
          pixelPoly= Polygon(np.array([[x1+(c*horizontaldiff_x),y1+(r*verticaldiff_y)],[x1+((c+1)*horizontaldiff_x),y1+(r*verticaldiff_y)],[x1+((c+1)*horizontaldiff_x),y1+((r+1)*verticaldiff_y)],[x1+((c)*horizontaldiff_x),y1+((r+1)*verticaldiff_y)]]))
          if point.within(pixelPoly):
            yd[r][c]+=X_2001[i][2]
            countyd[r][c]+=1
  for k in range(0,int(nrow),1):
    for m in range(0,int(ncol),1):
      if countyd[k][m]!=0:
        ans[k][m]=yd[k][m]/countyd[k][m]
  pixelYield_2001.append(ans)
```
